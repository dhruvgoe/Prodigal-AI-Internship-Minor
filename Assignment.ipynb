```python
import pandas as pd
from bs4 import BeautifulSoup as bs
import requests
import re
```

```python
url = "https://gov.optimism.io"
page = requests.get(url)
content = bs(page.content, "html.parser")
```

```python
content
```

# Task 1
## Scraped all the categories name link and description and saved into a csv

```python
categories = content.find_all("td", class_="category")
```

```python
categories
```

```python
category_names = [category.find("a").text for category in categories]
print(category_names)
```

```python
category_links = [category.find("a")["href"] for category in categories]
print(category_links)
```

```python
category_descriptions = [category.find("div", itemprop="description").text for category in categories]
print(category_descriptions)
```

```python
df = pd.DataFrame({
    "Category Name": category_names,
    "Category Link": category_links,
    "Category Description": category_descriptions
})
df.head()
```

```python
df = df.apply(lambda x: x.str.replace("\n", ""))
df = df.apply(lambda x: x.str.replace("\t", ""))
df = df.apply(lambda x: x.str.replace("\r", ""))
df = df.apply(lambda x: x.str.replace("\xa0", ""))
df.head()
```

```python
df
```

```python
df.to_csv("Categories.csv", index=False)
```

# Task 2
## Scraped all the articles name, link, date of posting and stored in respective csv files for the categories 

```python
def clean_category_name(category_name):
    first_word = re.match(r'\w+', category_name)
    return first_word.group(0) if first_word else "category"

def scrape_category_data(category_name, category_link):
    response = requests.get(category_link)
    data = bs(response.content, 'html.parser')
    articles = data.find_all('tr', class_='topic-list-item')
    
    data = []
    for article in articles:
        article_name_tag = article.find('a', class_='title')
        article_name = article_name_tag.get_text(strip=True) if article_name_tag else "N/A"
        
        article_link = article_name_tag['href'] if article_name_tag else "N/A"
        
        poster_tag = article.find('td', class_='posters').find('a')
        article_poster = poster_tag.get('aria-label', 'N/A').split(' -')[0] if poster_tag else "N/A"
        
        article_date = article.find_all('td')[-1].get_text(strip=True) if article.find_all('td') else "N/A"
        
        data.append({
            'Article Name': article_name,
            'Article Link': article_link,
            'Article Poster': article_poster,
            'Article Date': article_date
        })
    
    df_articles = pd.DataFrame(data)
    
    clean_name = clean_category_name(category_name)
    
    file_name = f"{clean_name}_data.csv"
    df_articles.to_csv(file_name, index=False)
    print(f"Data for category '{category_name}' saved to {file_name}.")

for index, row in df.iterrows():
    category_name = row['Category Name']
    category_link = row['Category Link']
    
    if not category_link.startswith("http"):
        category_link = f"https://gov.optimism.io{category_link}"
    
    scrape_category_data(category_name, category_link)
```
